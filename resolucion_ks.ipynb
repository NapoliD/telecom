```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```
datos = pd.read_csv('ejercico2_t_3.csv', sep= ',')
datos
```

```
malos=datos['bads'].sum()
buenos=datos['goods'].sum()
```

```

datos['odds'] = (datos.goods / datos.bads).apply('{0:.2f}'.format)
datos
```

```

datos['bad_rate'] = (datos.bads / datos.total).apply('{0:.2%}'.format)
datos
```

```
 # CALCULATE KS STATISTIC
 
datos['ks'] = np.round(((datos.bads / malos).cumsum() - (datos.goods / buenos).cumsum()), 4) * 100
 
```

```

 # DEFINE A FUNCTION TO FLAG MAX KS
flag = lambda x: '<----' if x == datos.ks.max() else ''
 
 # FLAG OUT MAX KS
 
datos['max_ks'] = datos.ks.apply(flag)
 
datos
```
